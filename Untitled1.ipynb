```python
import os


if not os.path.exists('plots'):
    os.makedirs('plots')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('netflix_titles.csv')




plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type')
plt.title('Movies vs TV Shows on Netflix')
plt.savefig('plots/movies_vs_tvshows.png')
plt.show()


df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind='bar', figsize=(10,6))
plt.title('Content Added to Netflix by Year')
plt.savefig('plots/content_by_year.png')
plt.show()


top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='barh', figsize=(8,5))
plt.title('Top 10 Countries Producing Netflix Content')
plt.savefig('plots/top_countries.png')
plt.show()
```
